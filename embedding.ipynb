# Embedding Layer 

```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
### sentences

sent = [
    'the glass of milk',
    'the glass of juice',
    'the cup of tea',
    'I am a good boy',
    'I am a good developer',
    'understand the meaning of words',
    'your videos are good'
]
```

```python
sent
```

```python
## next we need to define the vocabulary size (number of unique tokens)

voc_size = 10000 
```

```python
##one hot representation

one_hot_rep = [one_hot(words,voc_size)for words in sent]
one_hot_rep
```

```python
## word embedding representation

from tensorflow.keras.layers import Embedding
#from tensorflow.keras.processing.sequence import pad_sequences
from tensorflow.keras.utils import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
import numpy as np
```

```python
sent_length = 8
embedded_docs = pad_sequences(one_hot_rep,padding='pre',maxlen=sent_length)

#to get the zeros after the numbers we can use post
# embedded_docs = pad_sequences(one_hot_rep,padding='post',maxlen=sent_length)

print(embedded_docs)
```

```python
## feature representation

dim=10   # here we can use any number of dimensions in the  word embeddings and each word is mapped to 10 dimensianl vecotr
```

```python
model=Sequential() # define the model 
model.add(Embedding(voc_size,dim,input_length=sent_length))
model.compile('adam','mse') # here I am using adam optimizer and loss function mse to train the model



```

```python
model.build(input_shape=(None,8)) # tp build the model expicity if needed 
model.summary() # print the model 
```

```python
model.predict(embedded_docs)
```
